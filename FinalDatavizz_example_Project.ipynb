# Cleaning Data

## ตรวจสอบข้อมูล

```python
alldb_interest  = alldb[['user','RowKey','date','ov','phone','site','time','urine','PartitionKey','title','age']] #เลือกเฉพาะคอลัมที่สนใจ
alldb_interest
```

```python
alldb_interest.info() #ตรวจสอบข้อมูล
```

```python
alldb_interest.isnull().sum()
```

##แก้ค่า Missing และเปลี่ยน type ของข้อมูล

### คอลัม User

```python
alldb_interest= alldb_interest[(alldb_interest['user'].astype(str).str.isdigit()) & (alldb_interest['user'].str.len() == 5)] 
alldb_interest
```

```python
alldb_interest['user'] = pd.to_numeric(alldb_interest['user'], errors='coerce' )
```

```python
alldb_interest['user'] = alldb_interest['user'].astype('Int64')
```

```python
##set(alldb_interest['user'])
```

```python
alldb_interest[['user']] = alldb_interest[['user']].fillna(value ={ 'user' : 'ไม่ระบุ'}) #แทนค่าว่างด้วยคำว่า ไม่ระบุ
```
